```python
# Use a pipeline as a high-level helper
from transformers import pipeline
import torch
import gradio as gr
import json

pipe = pipeline("translation", model="facebook/nllb-200-distilled-600M")

with open('languages.json', 'r', encoding='utf-8') as f:
    data = json.load(f)

# Extract list of languages for dropdowns
language_names = [entry["language"] for entry in data]

def find_flores_code(language_name):
    for entry in data:
        if entry['language'].lower() == language_name.lower():
            return entry['code']
    return "Language not found."

def translate_text(text, source_language, target_language):
    source_code = find_flores_code(source_language)
    target_code = find_flores_code(target_language)

    translated_text = pipe(text,
                           src_lang=source_code,
                           tgt_lang=target_code)[0]['translation_text']
    return translated_text



# Gradio UI

gr.Interface(
    fn=translate_text,
    inputs=[
        gr.Textbox(lines=2, label="Enter Sentence to be Translated"),
        gr.Dropdown(choices=language_names, label="Select Source Language"),
        gr.Dropdown(choices=language_names, label="Select Destination Language")
    ],
    outputs=gr.Textbox(label="Translated Sentence"),
    title="Translator",
    description="Paste your sentence(s) here. This demo uses the facebook/nllb-200-distilled-600M model."
).launch()
```
